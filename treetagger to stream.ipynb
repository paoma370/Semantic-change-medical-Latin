```python
# Function to process TreeTagger output
def process_treetagger_output(file_path):
    lemmatized_sentences = []
    with open(file_path, 'r') as f:
        sentence = []
        for line in f:
            if line.strip():
                word, pos, lemma = line.strip().split('\t')
                sentence.append(lemma)
            else:
                lemmatized_sentences.append(' '.join(sentence))
                sentence = []
        # Add the last sentence if any
        if sentence:
            lemmatized_sentences.append(' '.join(sentence))
    return ' .\n'.join(lemmatized_sentences) + ' .'

# Process the TreeTagger output and save the lemmatized text to a file
lemmatized_text = process_treetagger_output('test-tagged.txt')
with open('test-ready.txt', 'w') as f:
    f.write(lemmatized_text)

print(lemmatized_text)
```
